```python
red_mask = cv2.inRange(image,(0,0,30),(80,80,255))
red_mask = red_mask.astype(bool)
green_mask = cv2.inRange(image,(0,30,0),(80,255,80))
green_mask = green_mask.astype(bool)

blue_mask = cv2.inRange(image,(30,0,0),(255,80,80))
blue_mask = blue_mask.astype(bool)

yellow_mask = cv2.inRange(image , (0,30,30) , (80,255,255))
yellow_mask = yellow_mask.astype(bool)

plt.figure()

```
